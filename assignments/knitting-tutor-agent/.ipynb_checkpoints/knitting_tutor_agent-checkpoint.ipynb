# KnitCoach: Collaborative Knitting Tutor Agent

## 과제 체크리스트

- Theme: 교육과 학습
- agent 설계에는 이름, 목적, 최소 3개의 핵심 기능, graph diagram을 포함합니다.
- 기본 구현은 LangGraph를 사용합니다.
- state는 custom `TypedDict`로 정의합니다.
- 최소 2개의 working node를 구현합니다.
- graph는 연결되어 있고 실행 가능해야 합니다.
- 설계 문서와 code를 이 Jupyter Notebook에 함께 포함합니다.


## 1단계: Agent 설계

**이름:** KnitCoach

**목적:** KnitCoach는 뜨개질 학습자가 코바늘(crochet)과 대바늘(needle knitting)을 구분하고, 완성품/도안/설명글에서 필요한 기법을 찾아 학습하도록 돕는 협업형 AI tutor agent입니다. 학습자의 수준과 목표를 진단하고, 도안 변환, 단계별 lesson, quiz, feynman feedback, practice plan을 제공합니다.

**핵심 기능:**

1. 학습자와 입력 분류: 질문, local image path, 도안 텍스트, 설명글, quiz 요청, feynman 요청을 구분합니다.
2. Artifact 분석: 완성품 이미지 경로나 도안 텍스트에서 코바늘/대바늘 여부와 기법을 추론합니다.
3. Pattern 변환: 도안 -> 초보자 설명글, 설명글 -> row-by-row 도안, 설명글 -> chart spec 변환을 수행합니다.
4. 맞춤 튜터링: 분석된 기법을 기준으로 teacher agent가 단계별 lesson을 제공합니다.
5. 학습 확인: quiz agent와 feynman agent가 이해도를 점검하고 feedback을 제공합니다.
6. Practice 설계: 학습자가 바로 수행할 수 있는 10-20분 practice plan과 next action을 생성합니다.

**학습 카테고리:**

- 코바늘(crochet): 사슬뜨기(chain stitch), 짧은뜨기(single crochet), 긴뜨기(double crochet), 빼뜨기(slip stitch), 원형뜨기, 모티브.
- 대바늘(needle knitting): 코 잡기(casting on), 겉뜨기(knit stitch), 안뜨기(purl stitch), 메리야스뜨기(stockinette), 고무뜨기(ribbing), 게이지(gauge), 코 막음(binding off).


## 그래프 구조

```mermaid
graph TD
  START --> classifier_agent
  classifier_agent --> router
  router --> artifact_analyzer_agent
  artifact_analyzer_agent --> pattern_converter_agent
  pattern_converter_agent --> chart_generator_agent
  chart_generator_agent --> technique_planner_agent
  technique_planner_agent --> teacher_agent
  teacher_agent --> quiz_agent
  quiz_agent --> feynman_agent
  feynman_agent --> evaluator_agent
  evaluator_agent --> practice_planner_agent
  practice_planner_agent --> check_understanding_agent
  check_understanding_agent --> END
```

- `classifier_agent`: 입력 유형, intent, 학습자 수준, artifact source를 분류합니다.
- `router`: 현재 요청을 담당할 specialist agent를 `active_agent`에 기록합니다.
- `artifact_analyzer_agent`: image path, link, 도안 텍스트, 설명글을 분석해 tool type과 기법을 추론합니다.
- `pattern_converter_agent`: 도안 -> 설명글, 설명글 -> 도안, 설명글 -> chart spec 변환을 수행합니다.
- `chart_generator_agent`: chart spec과 chart image path placeholder를 생성합니다.
- `technique_planner_agent`: 분석된 기법을 학습 순서로 정리합니다.
- `teacher_agent`: 현재 기법에 대한 단계별 lesson을 생성합니다.
- `quiz_agent`: quiz 요청일 때 quiz item과 feedback을 생성합니다.
- `feynman_agent`: feynman 요청일 때 학습자 설명에 대한 feedback을 생성합니다.
- `evaluator_agent`: specialist output을 요약해 다음 route 판단에 쓸 수 있게 정리합니다.
- `practice_planner_agent`: 10-20분 practice plan과 next action을 생성합니다.
- `check_understanding_agent`: 다음 턴으로 이어질 understanding check를 생성합니다.

v1에서는 notebook 실행 안정성을 위해 모든 node를 순차 실행하고, 각 node가 `intent`와 `active_agent`를 보고 필요한 작업만 수행합니다.


## 2단계: Basic LangGraph Build


````python
import os
from typing import Annotated, Literal, TypedDict

from dotenv import load_dotenv
from langchain_core.messages import AIMessage, BaseMessage, HumanMessage, SystemMessage
from langgraph.graph import END, START, StateGraph, add_messages

load_dotenv()

MODEL_NAME = os.getenv("KNITTING_AGENT_MODEL", "gpt-4.1-mini")
HAS_OPENAI_KEY = bool(os.getenv("OPENAI_API_KEY"))
````

````python
class KnitCoachState(TypedDict):
    messages: Annotated[list[BaseMessage], add_messages]
    learner_profile: dict
    input_type: Literal["text", "image_path", "pattern_text", "explanation_text", "link", "feedback"]
    intent: Literal["analyze_artifact", "convert_pattern", "learn_technique", "quiz", "feynman", "revise_output"]
    artifact_source: str
    artifact_analysis: str
    tool_type: Literal["crochet", "needle_knitting", "unknown"]
    pattern_type: Literal["photo", "finished_object", "written_pattern", "symbol_chart", "diagram_chart", "unknown"]
    detected_techniques: list[str]
    active_agent: str
    conversion_mode: Literal["pattern_to_explanation", "explanation_to_pattern", "explanation_to_chart", "tutor_only"]
    generated_explanation: str
    generated_pattern: str
    chart_spec: dict
    chart_image_path: str
    user_feedback: str
    quiz_items: list[dict]
    quiz_feedback: str
    feynman_prompt: str
    feynman_feedback: str
    goal: str
    current_skill: str
    blockers: list[str]
    learning_path: Literal["crochet_beginner", "needle_knitting_beginner", "diagnose_tool"]
    curriculum_stage: str
    lesson_summary: str
    practice_plan: str
    understanding_check: str
    next_action: str
````

````python
CROCHET_KEYWORDS = {
    "코바늘", "crochet", "사슬", "사슬뜨기", "짧은뜨기", "긴뜨기", "빼뜨기", "모티브", "원형뜨기", "컵받침", "ch", "sc", "dc", "sl st"
}
NEEDLE_KEYWORDS = {
    "대바늘", "knit", "knitting", "겉뜨기", "안뜨기", "메리야스", "고무뜨기", "코잡기", "코 잡기", "목도리", "k2tog", "yo", "purl"
}
PATTERN_KEYWORDS = {"ch", "sc", "dc", "sl st", "row", "round", "repeat", "k", "p", "yo", "k2tog"}
BLOCKER_KEYWORDS = {
    "늘어나": "코가 늘어남",
    "빠져": "코가 빠짐",
    "조여": "장력이 너무 조임",
    "헐거": "장력이 너무 헐거움",
    "말려": "편물이 말림",
    "단 수": "단 수가 맞지 않음",
    "어려": "기술 이해가 어려움",
}

CROCHET_CURRICULUM = {
    "사슬뜨기": "1. 사슬뜨기로 일정한 foundation 만들기",
    "짧은뜨기": "2. 짧은뜨기로 작은 사각형 만들기",
    "긴뜨기": "3. 긴뜨기로 높이와 장력 비교하기",
    "원형뜨기": "4. 원형뜨기에서 늘림 위치 익히기",
}
NEEDLE_CURRICULUM = {
    "코 잡기": "1. 코를 일정한 간격으로 잡기",
    "겉뜨기": "2. 겉뜨기로 garter stitch 만들기",
    "안뜨기": "3. 안뜨기와 겉뜨기 방향 구분하기",
    "게이지": "4. 게이지를 재고 바늘 크기 조정하기",
}


def latest_user_text(state: KnitCoachState) -> str:
    for message in reversed(state["messages"]):
        if isinstance(message, HumanMessage):
            return str(message.content)
    return ""


def detect_tool_type(text: str) -> Literal["crochet", "needle_knitting", "unknown"]:
    lowered = text.lower()
    has_crochet = any(keyword.lower() in lowered for keyword in CROCHET_KEYWORDS)
    has_needle = any(keyword.lower() in lowered for keyword in NEEDLE_KEYWORDS)
    if has_crochet and not has_needle:
        return "crochet"
    if has_needle and not has_crochet:
        return "needle_knitting"
    return "unknown"


def detect_input_type(text: str) -> Literal["text", "image_path", "pattern_text", "explanation_text", "link", "feedback"]:
    lowered = text.lower()
    if "http://" in lowered or "https://" in lowered:
        return "link"
    if any(ext in lowered for ext in [".png", ".jpg", ".jpeg", ".webp"]):
        return "image_path"
    if any(keyword in lowered for keyword in PATTERN_KEYWORDS):
        return "pattern_text"
    if any(word in text for word in ["수정", "바꿔", "더 잘", "피드백 반영"]):
        return "feedback"
    if any(word in text for word in ["도안으로", "차트", "chart"]):
        return "explanation_text"
    return "text"


def detect_intent(text: str, input_type: str) -> Literal["analyze_artifact", "convert_pattern", "learn_technique", "quiz", "feynman", "revise_output"]:
    if any(word in text for word in ["퀴즈", "quiz", "문제"]):
        return "quiz"
    if any(word in text for word in ["파인만", "feynman", "설명해볼게", "내 말로"]):
        return "feynman"
    if input_type == "feedback":
        return "revise_output"
    if input_type in {"image_path", "link"} or any(word in text for word in ["분석", "완성품", "사진", "이미지"]):
        return "analyze_artifact"
    if input_type in {"pattern_text", "explanation_text"} or any(word in text for word in ["설명글", "도안으로", "변환", "chart"]):
        return "convert_pattern"
    return "learn_technique"


def detect_current_skill(text: str, tool_type: str) -> str:
    skill_keywords = [
        "사슬뜨기", "짧은뜨기", "긴뜨기", "빼뜨기", "원형뜨기", "모티브",
        "코 잡기", "코잡기", "겉뜨기", "안뜨기", "메리야스", "고무뜨기", "게이지", "코 막음",
    ]
    for keyword in skill_keywords:
        if keyword in text:
            return "코 잡기" if keyword == "코잡기" else keyword
    lowered = text.lower()
    if "sc" in lowered:
        return "짧은뜨기"
    if "dc" in lowered:
        return "긴뜨기"
    if "ch" in lowered:
        return "사슬뜨기"
    if "knit" in lowered or " k" in lowered:
        return "겉뜨기"
    if tool_type == "crochet":
        return "사슬뜨기"
    if tool_type == "needle_knitting":
        return "코 잡기"
    return "도구 선택"


def detect_goal(text: str, tool_type: str) -> str:
    if "목도리" in text:
        return "목도리 만들기"
    if "컵받침" in text:
        return "컵받침 만들기"
    if "비니" in text or "모자" in text:
        return "모자 만들기"
    if tool_type == "crochet":
        return "코바늘 기초 익히기"
    if tool_type == "needle_knitting":
        return "대바늘 기초 익히기"
    return "나에게 맞는 뜨개 도구와 첫 프로젝트 정하기"


def detect_blockers(text: str) -> list[str]:
    return [label for keyword, label in BLOCKER_KEYWORDS.items() if keyword in text]


def detect_techniques(text: str, tool_type: str, current_skill: str) -> list[str]:
    techniques = []
    lowered = text.lower()
    technique_map = {
        "ch": "사슬뜨기(chain stitch)",
        "sc": "짧은뜨기(single crochet)",
        "dc": "긴뜨기(double crochet)",
        "sl st": "빼뜨기(slip stitch)",
        "겉뜨기": "겉뜨기(knit stitch)",
        "안뜨기": "안뜨기(purl stitch)",
        "k2tog": "왼코/오른코 줄임(k2tog)",
        "yo": "바늘비우기(yarn over)",
    }
    for keyword, label in technique_map.items():
        if keyword in lowered or keyword in text:
            techniques.append(label)
    if not techniques and current_skill != "도구 선택":
        techniques.append(current_skill)
    if not techniques and tool_type == "crochet":
        techniques.append("사슬뜨기(chain stitch)")
    if not techniques and tool_type == "needle_knitting":
        techniques.append("코 잡기(casting on)")
    return techniques
````

````python
def classifier_agent(state: KnitCoachState) -> dict:
    text = latest_user_text(state)
    input_type = detect_input_type(text)
    tool_type = detect_tool_type(text)
    current_skill = detect_current_skill(text, tool_type)
    intent = detect_intent(text, input_type)
    learner_level = "beginner" if any(word in text for word in ["처음", "초보", "입문"]) else "unknown"
    learner_profile = {
        "level": learner_level,
        "preferred_style": "hands_on" if "연습" in text else "unknown",
        "goal": detect_goal(text, tool_type),
    }

    return {
        "input_type": input_type,
        "intent": intent,
        "artifact_source": text,
        "tool_type": tool_type,
        "learner_profile": learner_profile,
        "goal": learner_profile["goal"],
        "current_skill": current_skill,
        "blockers": detect_blockers(text),
        "detected_techniques": detect_techniques(text, tool_type, current_skill),
        "user_feedback": text if intent == "revise_output" else "",
    }


def router(state: KnitCoachState) -> dict:
    intent_to_agent = {
        "analyze_artifact": "artifact_analyzer_agent",
        "convert_pattern": "pattern_converter_agent",
        "learn_technique": "teacher_agent",
        "quiz": "quiz_agent",
        "feynman": "feynman_agent",
        "revise_output": "user_feedback_agent",
    }
    return {"active_agent": intent_to_agent.get(state["intent"], "teacher_agent")}


def artifact_analyzer_agent(state: KnitCoachState) -> dict:
    if state["input_type"] not in {"image_path", "link", "pattern_text"} and state["intent"] != "analyze_artifact":
        return {}

    if state["input_type"] == "image_path":
        analysis = (
            "local image path가 감지되었습니다. v1 fallback에서는 이미지 자체 vision 분석 대신 "
            "파일명과 사용자 설명을 기준으로 코바늘/대바늘 가능성과 기법을 추론합니다."
        )
        pattern_type = "photo"
    elif state["input_type"] == "link":
        analysis = "link input이 감지되었습니다. v1에서는 실제 fetch 없이 URL source만 기록합니다."
        pattern_type = "unknown"
    elif state["input_type"] == "pattern_text":
        analysis = "written pattern이 감지되었습니다. abbreviation과 row/round 표현을 기준으로 기법을 추론합니다."
        pattern_type = "written_pattern"
    else:
        analysis = "사용자 설명에서 완성품 또는 기법 학습 의도를 분석했습니다."
        pattern_type = "finished_object"

    tool_type = state["tool_type"]
    if tool_type == "unknown" and any("crochet" in t or "사슬" in t or "짧은" in t for t in state["detected_techniques"]):
        tool_type = "crochet"
    elif tool_type == "unknown" and any("knit" in t or "겉뜨기" in t or "안뜨기" in t for t in state["detected_techniques"]):
        tool_type = "needle_knitting"

    return {
        "artifact_analysis": analysis,
        "pattern_type": pattern_type,
        "tool_type": tool_type,
    }


def pattern_converter_agent(state: KnitCoachState) -> dict:
    text = latest_user_text(state)
    lowered = text.lower()
    should_convert = state["intent"] in {"convert_pattern", "revise_output"} or state["input_type"] in {"pattern_text", "explanation_text"}
    if not should_convert:
        return {}

    if state["input_type"] == "pattern_text" or any(k in lowered for k in ["ch", "sc", "dc", "row"]):
        generated_explanation = (
            "도안 해석: 먼저 사슬뜨기(ch)로 시작 코를 만들고, 각 row에서 짧은뜨기(sc)를 반복합니다. "
            "초보자는 각 row가 끝날 때 코 수가 유지되는지 확인하세요."
        )
        generated_pattern = text
        conversion_mode = "pattern_to_explanation"
    else:
        generated_pattern = "Row 1: ch 12.\nRow 2-6: sc across, ch 1, turn.\nFinish: 마지막 코를 정리하고 실 끝을 숨깁니다."
        generated_explanation = "설명글을 코바늘 컵받침용 row-by-row pattern으로 변환했습니다."
        conversion_mode = "explanation_to_chart" if "chart" in lowered or "이미지" in text or "도안" in text else "explanation_to_pattern"

    chart_spec = {
        "title": "Beginner Crochet Coaster Chart",
        "tool_type": "crochet" if state["tool_type"] != "needle_knitting" else "needle_knitting",
        "rows": [
            ["ch"] * 12,
            ["sc"] * 12,
            ["sc"] * 12,
            ["sc"] * 12,
            ["sc"] * 12,
            ["sc"] * 12,
        ],
        "legend": {"ch": "사슬뜨기", "sc": "짧은뜨기"},
    }
    return {
        "conversion_mode": conversion_mode,
        "generated_explanation": generated_explanation,
        "generated_pattern": generated_pattern,
        "chart_spec": chart_spec,
        "tool_type": chart_spec["tool_type"],
        "detected_techniques": state["detected_techniques"] or ["사슬뜨기(chain stitch)", "짧은뜨기(single crochet)"],
    }


def chart_generator_agent(state: KnitCoachState) -> dict:
    if not state["chart_spec"]:
        return {}
    chart_image_path = "outputs/knitcoach_chart_preview.png"
    return {"chart_image_path": chart_image_path}


def technique_planner_agent(state: KnitCoachState) -> dict:
    tool_type = state["tool_type"]
    current_skill = state["current_skill"]
    if tool_type == "crochet":
        learning_path = "crochet_beginner"
        curriculum_stage = CROCHET_CURRICULUM.get(current_skill, CROCHET_CURRICULUM["사슬뜨기"])
    elif tool_type == "needle_knitting":
        learning_path = "needle_knitting_beginner"
        curriculum_stage = NEEDLE_CURRICULUM.get(current_skill, NEEDLE_CURRICULUM["코 잡기"])
    else:
        learning_path = "diagnose_tool"
        curriculum_stage = "0. 목표 작품과 보유 도구를 확인해 학습 path 결정하기"
    return {"learning_path": learning_path, "curriculum_stage": curriculum_stage}
````

````python
def teacher_agent(state: KnitCoachState) -> dict:
    techniques = ", ".join(state["detected_techniques"]) if state["detected_techniques"] else state["current_skill"]
    blockers = ", ".join(state["blockers"]) if state["blockers"] else "아직 뚜렷한 막힘은 없음"

    if state["learning_path"] == "crochet_beginner":
        lesson = (
            f"코바늘(crochet) path입니다. 현재 curriculum stage는 '{state['curriculum_stage']}'입니다. "
            f"오늘 focus technique은 {techniques}입니다. 실을 너무 세게 당기지 말고, 바늘이 코 안을 "
            f"편하게 지나갈 정도의 장력을 유지하세요. 현재 blocker: {blockers}."
        )
    elif state["learning_path"] == "needle_knitting_beginner":
        lesson = (
            f"대바늘(needle knitting) path입니다. 현재 curriculum stage는 '{state['curriculum_stage']}'입니다. "
            f"오늘 focus technique은 {techniques}입니다. 바늘 끝이 아니라 몸통 부분에서 코 크기를 맞추고, "
            f"매 단 끝에서 코 수를 확인하세요. 현재 blocker: {blockers}."
        )
    else:
        lesson = (
            "아직 tool path가 확정되지 않았습니다. 작은 소품이나 모티브를 빨리 만들고 싶다면 코바늘, "
            "목도리처럼 평평한 편물을 반복하고 싶다면 대바늘이 적합합니다."
        )

    if state["generated_explanation"]:
        lesson += f"\n\n변환된 설명: {state['generated_explanation']}"
    return {"lesson_summary": lesson}


def quiz_agent(state: KnitCoachState) -> dict:
    if state["intent"] != "quiz":
        return {}
    technique = state["detected_techniques"][0] if state["detected_techniques"] else state["current_skill"]
    quiz_items = [
        {
            "question": f"{technique}을 연습할 때 매 단마다 확인해야 하는 것은 무엇인가요?",
            "choices": ["코 수", "실 색상", "가위 위치", "책상 높이"],
            "answer": "코 수",
            "explanation": "초보자는 코가 늘거나 줄기 쉬우므로 매 단마다 코 수를 확인해야 합니다.",
        }
    ]
    return {"quiz_items": quiz_items, "quiz_feedback": "1문항 quiz를 생성했습니다. 답을 고르면 해설과 함께 복습할 수 있습니다."}


def feynman_agent(state: KnitCoachState) -> dict:
    if state["intent"] != "feynman":
        return {}
    feedback = (
        "파인만 feedback: 설명에 '어떤 손동작을 먼저 하는지', '코를 어디에 넣는지', "
        "'완료 후 무엇을 확인하는지'가 들어가면 더 명확해집니다. 초등학생에게 말하듯 한 문장씩 짧게 다시 설명해 보세요."
    )
    return {
        "feynman_prompt": "방금 배운 기법을 도구를 처음 잡은 친구에게 설명하듯 말해 보세요.",
        "feynman_feedback": feedback,
    }


def evaluator_agent(state: KnitCoachState) -> dict:
    notes = []
    if state["quiz_feedback"]:
        notes.append(state["quiz_feedback"])
    if state["feynman_feedback"]:
        notes.append(state["feynman_feedback"])
    if state["artifact_analysis"]:
        notes.append(state["artifact_analysis"])
    if not notes:
        return {}
    return {"artifact_analysis": "\n".join(notes)}


def practice_planner_agent(state: KnitCoachState) -> dict:
    if state["learning_path"] == "crochet_beginner":
        practice_plan = "사슬 12코를 만들고 짧은뜨기 5단을 떠 보세요. 각 단 끝에서 코 수를 확인합니다."
        next_action = "샘플을 완성한 뒤 가장 조이거나 헐거운 단을 표시하세요."
    elif state["learning_path"] == "needle_knitting_beginner":
        practice_plan = "15코를 잡고 겉뜨기 6단을 떠 보세요. 매 단 끝에서 코 수가 15개인지 확인합니다."
        next_action = "코가 늘어난 단이 몇 번째 단인지 적어 보세요."
    else:
        practice_plan = "가지고 있는 바늘 종류, 만들고 싶은 첫 작품, 하루 연습 가능 시간을 적어 보세요."
        next_action = "코바늘/대바늘 중 가지고 있는 도구와 만들고 싶은 작품을 한 문장으로 정리하세요."
    return {"practice_plan": practice_plan, "next_action": next_action}


def check_understanding_agent(state: KnitCoachState) -> dict:
    if state["intent"] == "quiz" and state["quiz_items"]:
        understanding_check = state["quiz_items"][0]["question"]
    elif state["intent"] == "feynman":
        understanding_check = "방금 배운 기법을 한 문장으로 다시 설명한다면 어떻게 말하겠어요?"
    elif state["generated_pattern"]:
        understanding_check = "생성된 pattern에서 반복되는 row는 몇 개인가요?"
    elif state["learning_path"] == "crochet_beginner":
        understanding_check = "짧은뜨기를 연습할 때 매 단마다 무엇을 확인해야 할까요?"
    elif state["learning_path"] == "needle_knitting_beginner":
        understanding_check = "겉뜨기 연습 중 코가 늘어나는지 확인하려면 언제 코 수를 세어야 할까요?"
    else:
        understanding_check = "작은 소품과 평평한 편물 중 어떤 결과물을 먼저 만들고 싶나요?"

    sections = [
        f"### 담당 agent\n{state['active_agent']}",
        f"### 분석된 기법\n{', '.join(state['detected_techniques']) if state['detected_techniques'] else '아직 확정되지 않음'}",
        f"### Lesson\n{state['lesson_summary']}",
    ]
    if state["generated_pattern"]:
        sections.append(f"### Generated pattern\n{state['generated_pattern']}")
    if state["chart_spec"]:
        sections.append(f"### Chart spec\n{state['chart_spec']}")
        sections.append(f"### Chart image path\n{state['chart_image_path']}")
    if state["quiz_items"]:
        sections.append(f"### Quiz\n{state['quiz_items'][0]}")
    if state["feynman_feedback"]:
        sections.append(f"### Feynman feedback\n{state['feynman_feedback']}")
    sections.extend([
        f"### Practice plan\n{state['practice_plan']}",
        f"### Next action\n{state['next_action']}",
        f"### Understanding check\n{understanding_check}",
    ])

    return {
        "understanding_check": understanding_check,
        "messages": [AIMessage(content="\n\n".join(sections))],
    }
````

````python
graph_builder = StateGraph(KnitCoachState)

graph_builder.add_node("classifier_agent", classifier_agent)
graph_builder.add_node("router", router)
graph_builder.add_node("artifact_analyzer_agent", artifact_analyzer_agent)
graph_builder.add_node("pattern_converter_agent", pattern_converter_agent)
graph_builder.add_node("chart_generator_agent", chart_generator_agent)
graph_builder.add_node("technique_planner_agent", technique_planner_agent)
graph_builder.add_node("teacher_agent", teacher_agent)
graph_builder.add_node("quiz_agent", quiz_agent)
graph_builder.add_node("feynman_agent", feynman_agent)
graph_builder.add_node("evaluator_agent", evaluator_agent)
graph_builder.add_node("practice_planner_agent", practice_planner_agent)
graph_builder.add_node("check_understanding_agent", check_understanding_agent)

graph_builder.add_edge(START, "classifier_agent")
graph_builder.add_edge("classifier_agent", "router")
graph_builder.add_edge("router", "artifact_analyzer_agent")
graph_builder.add_edge("artifact_analyzer_agent", "pattern_converter_agent")
graph_builder.add_edge("pattern_converter_agent", "chart_generator_agent")
graph_builder.add_edge("chart_generator_agent", "technique_planner_agent")
graph_builder.add_edge("technique_planner_agent", "teacher_agent")
graph_builder.add_edge("teacher_agent", "quiz_agent")
graph_builder.add_edge("quiz_agent", "feynman_agent")
graph_builder.add_edge("feynman_agent", "evaluator_agent")
graph_builder.add_edge("evaluator_agent", "practice_planner_agent")
graph_builder.add_edge("practice_planner_agent", "check_understanding_agent")
graph_builder.add_edge("check_understanding_agent", END)

knitcoach = graph_builder.compile()
````

````python
def initial_state(user_text: str) -> KnitCoachState:
    return {
        "messages": [HumanMessage(content=user_text)],
        "learner_profile": {},
        "input_type": "text",
        "intent": "learn_technique",
        "artifact_source": "",
        "artifact_analysis": "",
        "tool_type": "unknown",
        "pattern_type": "unknown",
        "detected_techniques": [],
        "active_agent": "",
        "conversion_mode": "tutor_only",
        "generated_explanation": "",
        "generated_pattern": "",
        "chart_spec": {},
        "chart_image_path": "",
        "user_feedback": "",
        "quiz_items": [],
        "quiz_feedback": "",
        "feynman_prompt": "",
        "feynman_feedback": "",
        "goal": "",
        "current_skill": "",
        "blockers": [],
        "learning_path": "diagnose_tool",
        "curriculum_stage": "",
        "lesson_summary": "",
        "practice_plan": "",
        "understanding_check": "",
        "next_action": "",
    }


def run_example(user_text: str) -> None:
    result = knitcoach.invoke(initial_state(user_text))
    print("USER:", user_text)
    print("INTENT:", result["intent"])
    print("ACTIVE_AGENT:", result["active_agent"])
    print("INPUT_TYPE:", result["input_type"])
    print("PATH:", result["learning_path"])
    print("CURRICULUM:", result["curriculum_stage"])
    print("TECHNIQUES:", result["detected_techniques"])
    print("ASSISTANT:", result["messages"][-1].content)
    print("-" * 80)
````

````python
run_example("뜨개질 처음인데 목도리를 만들고 싶어. 코바늘이랑 대바늘 중 뭐가 좋아?")
run_example("코바늘로 컵받침을 만들고 싶은데 사슬뜨기 다음에 뭘 해야 해?")
run_example("대바늘 겉뜨기를 하다 보면 코가 자꾸 늘어나.")
run_example("이 이미지 경로의 완성품이 코바늘인지 대바늘인지 분석하고 필요한 기법을 알려줘: samples/crochet-coaster.png")
run_example("ch 12, sc across for 5 rows 이 도안을 초보자 설명글로 바꿔줘.")
run_example("컵받침을 코바늘 짧은뜨기로 만드는 설명을 도안과 chart spec으로 바꿔줘.")
run_example("방금 배운 짧은뜨기로 퀴즈를 내줘.")
run_example("내가 짧은뜨기를 설명해볼게. 틀린 부분을 파인만 방식으로 피드백해줘.")
````
